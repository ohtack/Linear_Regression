```python
import pandas as pd
import numpy as np
from tensorflow import keras

import scipy as sp
import scipy.stats as stats  

import statsmodels.api as sm
from sklearn import linear_model
from statsmodels.formula.api import ols
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

```

```python
dataset_path = keras.utils.get_file("auto-mpg.data", "http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data")
dataset_path
```

```python
column_names = ['MPG','Cylinders','Displacement','Horsepower','Weight',
                'Acceleration', 'Model Year', 'Origin']
raw_dataset = pd.read_csv(dataset_path, names=column_names,
                      na_values = "?", comment='\t',
                      sep=" ", skipinitialspace=True)

X = raw_dataset.copy()

X.describe().transpose()
```

### Clean the data

```python
# Data contains unknown values

X.isna().sum()
```

```python
# drop unknown 

X = X.dropna()
```

```python
origin = X.pop('Origin')
```

```python
X['USA'] = (origin == 1)*1.0
X['Europe'] = (origin == 2)*1.0
X['Japan'] = (origin == 3)*1.0
X.tail()
```

```python
sns.pairplot(X[["MPG","Cylinders", "Displacement", "Weight", "Horsepower"]], diag_kind="kde")
```

```python
y = X['MPG']
X = X.iloc[:,1:]

print(X.shape)
print(y.shape)
```

```python
# Split the data into train and test

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 2021)
```

```python
X_train.describe().transpose()
```

```python
X_train.columns
```

```python
# Apply data into model

X_train = sm.add_constant(X_train) # beta0 value added (1.0 added)
model = sm.OLS(y_train, X_train, axis = 1)
model_trained = model.fit()
```

```python
# Check residuals are normally distributed 
#    

model_residuals = model_trained.resid

plt.rcParams['axes.unicode_minus'] = False
fig, ax = plt.subplots(1,1)
fig.set_figheight(10)
fig.set_figwidth(10)

sm.ProbPlot(model_residuals).qqplot(line = 's', color = '#1f77b4', ax = ax)
ax.title.set_text('QQ plot')
```

```python
# Check Homoskedasticity of varianceb
#   

model_fitted_y = model_trained.fittedvalues

fig, ax = plt.subplots(1, 1)
fig.set_figheight(4)
fig.set_figwidth(6)

sns.residplot(model_fitted_y, y_train, data = X_train, lowess = True, scatter_kws = {'alpha': 0.5},
             line_kws={'color': 'red'}, ax = ax)
ax.title.set_text('Residuals vs Fitted')
ax.set(xlabel = 'Fitted values', ylabel = 'Residuals')
```

### Model Evaluation

```python
# 3 metrics for model evaluation in regression:
# Mean Square Error(MSE)/Root Mean Square Error(RMSE)
# Mean Absolute Error(MAE)
# R Square/Adjusted R Square


# R Squred :  
# Coef : X     Y 
# P > [t] (p-value) : 0.05 ( )    
```

```python
X_test = sm.add_constant(X_test)
print(X_test)
```

```python
y_test_pred = model_trained.predict(X_test.drop(['Displacement', 'Acceleration','Cylinders'], axis = 1))
y_test_pred.head()
```

#### Mean Square Error (MSE)

$$ \frac{1}{n}\sum_{i=1}^n ({y_i - \hat{y}})^{2} $$

```python
print(mean_squared_error(y_test, y_test_pred))
```

#### Root Mean Square Error(RMSE)

$$ \sqrt{\frac{1}{n}\sum_{i=1}^n ({y_i - \hat{y}})^{2}} $$

```python
print(np.sqrt(mean_squared_error(y_test, y_test_pred)))
```

#### Mean Absolute Error(MAE)

$$ \frac{1}{n}\sum_{i=1}^n |{y_i - \hat{y}}| $$

```python
print(mean_absolute_error(y_test, y_test_pred))
```

#### Mean Absolute Percentage Error

$$ \frac{1}{n}\sum_{i=1}^n |\frac{y_i - \hat{y}}{y_i}| \times 100% $$

```python
def mean_absolute_percentage_error(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true-y_pred)/y_true))*100

print(mean_absolute_percentage_error(y_test, y_test_pred))
```

#### Coefficient of determination $$ R^2  = \frac {SSR}{SST} = 1 - \frac {SSE}{SST}$$

```python
print(r2_score(y_test, y_test_pred))
```
